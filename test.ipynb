```python
import time
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 크롬 드라이버 자동 설치 및 경로 설정
chromedriver_autoinstaller.install()

# 크롬 드라이버 실행 옵션 설정 (GUI 실행)
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--start-maximized")  # 최대화된 상태로 시작

# 드라이버 초기화 (헤드리스 모드 없이 실행)
driver = webdriver.Chrome(options=chrome_options)

# 웹사이트로 이동
url = "https://webresearch.pythonanywhere.com/"
driver.get(url)

wait = WebDriverWait(driver, 3)

def press_enter():
    """엔터 키 입력"""
    webdriver.ActionChains(driver).send_keys(Keys.ENTER).perform()
    print("엔터 입력 완료")
    time.sleep(0.5)

def click_button(button_id):
    """버튼 클릭"""
    button = wait.until(EC.element_to_be_clickable((By.ID, button_id)))
    button.click()
    print(f"버튼 {button_id} 클릭 완료")
    time.sleep(1)

def select_radio_button(value):
    """라디오 버튼 선택"""
    radio_button = wait.until(EC.element_to_be_clickable(
        (By.CSS_SELECTOR, f'input[name="jspsych-survey-multi-choice-response-0"][value="{value}"]')
    ))
    radio_button.click()
    print(f"라디오 버튼 {value} 선택 완료")
    time.sleep(4) ## 2~5초 사이 난수

# 순서대로 동작 수행
try:
    press_enter()
    click_button('jspsych-instructions-next')
    click_button('jspsych-instructions-next')
    click_button('jspsych-instructions-next')
    press_enter()

    select_radio_button('C')
    click_button('jspsych-survey-multi-choice-next')
    press_enter()

    press_enter()
    select_radio_button('F')
    click_button('jspsych-survey-multi-choice-next')
    press_enter()

    # 여러 라디오 버튼 반복 선택
    press_enter()
    for _ in range(8):
        select_radio_button('F')
        click_button('jspsych-survey-multi-choice-next')

finally:
    # 테스트 완료 후 브라우저 종료
    driver.quit()
```

```python
import time
import random
import os
import json
from joblib import Parallel, delayed
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from browsermobproxy import Server

# BrowserMob Proxy 경로 설정
BROWSERMOB_PATH = "C:\\Users\\PSID_PC_20\\Desktop\\PSID연구실\\Environment(작업환경_업무팁_이론정리)\\BrowserMob Proxy\\browsermob-proxy-2.1.4\\bin\\browsermob-proxy"
chromedriver_autoinstaller.install()

def start_proxy():
    """BrowserMob Proxy 서버 시작"""
    server = Server(BROWSERMOB_PATH)
    server.start()
    return server.create_proxy()

def user_simulation(user_id, total_users):
    """사용자 행동 시뮬레이션 함수"""
    print(f"사용자 {user_id}/{total_users} 시뮬레이션 시작")

    # BrowserMob Proxy 생성 및 설정
    proxy = start_proxy()
    proxy.new_har(f"user_{user_id}", options={'captureContent': True})

    # 크롬 드라이버 실행 옵션 설정
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument(f"--proxy-server={proxy.proxy}")
    chrome_options.add_argument("--start-maximized")

    driver = webdriver.Chrome(options=chrome_options)
    wait = WebDriverWait(driver, 5)

    try:
        # 웹사이트로 이동
        url = "http://webresearch.pythonanywhere.com/"
        driver.get(url)

        def press_enter():
            """엔터 키 입력"""
            webdriver.ActionChains(driver).send_keys(Keys.ENTER).perform()
            print(f"사용자 {user_id}: 엔터 입력 완료")
            time.sleep(random.uniform(0.5, 1.5))

        def click_button(button_id):
            """버튼 클릭"""
            button = wait.until(EC.element_to_be_clickable((By.ID, button_id)))
            button.click()
            print(f"사용자 {user_id}: 버튼 {button_id} 클릭 완료")
            time.sleep(random.uniform(1, 2))

        def select_radio_button(value):
            """라디오 버튼 선택"""
            radio_button = wait.until(EC.element_to_be_clickable(
                (By.CSS_SELECTOR, f'input[name="jspsych-survey-multi-choice-response-0"][value="{value}"]')
            ))
            radio_button.click()
            print(f"사용자 {user_id}: 라디오 버튼 {value} 선택 완료")
            time.sleep(random.uniform(2, 5))

        # 동작 수행
        press_enter()
        click_button('jspsych-instructions-next')
        click_button('jspsych-instructions-next')
        click_button('jspsych-instructions-next')
        press_enter()

        select_radio_button('C')
        click_button('jspsych-survey-multi-choice-next')
        press_enter()

        press_enter()
        select_radio_button('F')
        click_button('jspsych-survey-multi-choice-next')
        press_enter()
        press_enter()

        for _ in range(8):
            select_radio_button('F')
            click_button('jspsych-survey-multi-choice-next')

    finally:
        # HAR 파일 저장
        har_data = proxy.har
        with open(f"user_{user_id}_network_log.har", "w") as f:
            json.dump(har_data, f)

        # 브라우저 종료 및 프록시 서버 중지
        driver.quit()
        proxy.close()
        print(f"사용자 {user_id} 시뮬레이션 완료")

def run_parallel_simulations(num_users=40, num_cpus=2):
    """40명의 사용자 시뮬레이션을 병렬로 실행"""
    Parallel(n_jobs=num_cpus)(
        delayed(user_simulation)(user_id, num_users) for user_id in range(num_users)
    )

if __name__ == "__main__":
    run_parallel_simulations()
```

```python
import asyncio
import json
import random
import time
from playwright.async_api import async_playwright

async def user_simulation(user_id):
    """사용자 시뮬레이션 함수"""
    print(f"사용자 {user_id} 시뮬레이션 시작")

    async with async_playwright() as p:
        # 브라우저 실행
        browser = await p.chromium.launch(headless=False)
        context = await browser.new_context()
        page = await context.new_page()

        # 네트워크 요청 추적
        network_log = []

        def log_request(request):
            """HTTP 요청 로그"""
            network_log.append({
                "url": request.url,
                "method": request.method,
                "headers": request.headers
            })

        page.on("request", log_request)

        try:
            # 웹사이트 접속
            await page.goto("http://webresearch.pythonanywhere.com/")

            async def press_enter():
                """엔터 키 입력"""
                await page.keyboard.press("Enter")
                print(f"사용자 {user_id}: 엔터 입력 완료")
                await asyncio.sleep(random.uniform(0.5, 1.5))

            async def click_button(button_id):
                """버튼 클릭"""
                await page.click(f'#{button_id}')
                print(f"사용자 {user_id}: 버튼 {button_id} 클릭 완료")
                await asyncio.sleep(random.uniform(1, 2))

            async def select_radio_button(value):
                """라디오 버튼 선택"""
                await page.click(f'input[name="jspsych-survey-multi-choice-response-0"][value="{value}"]')
                print(f"사용자 {user_id}: 라디오 버튼 {value} 선택 완료")
                await asyncio.sleep(random.uniform(2, 5))

            # 동작 수행
            await press_enter()
            await click_button('jspsych-instructions-next')
            await click_button('jspsych-instructions-next')
            await click_button('jspsych-instructions-next')
            await press_enter()

            await select_radio_button('C')
            await click_button('jspsych-survey-multi-choice-next')
            await press_enter()

            await press_enter()
            await select_radio_button('F')
            await click_button('jspsych-survey-multi-choice-next')
            await press_enter()

            # 여러 라디오 버튼 반복 선택
            for _ in range(8):
                await select_radio_button('F')
                await click_button('jspsych-survey-multi-choice-next')

        finally:
            # 네트워크 로그 저장
            with open(f"user_{user_id}_network_log.json", "w") as f:
                json.dump(network_log, f)

            # 브라우저 종료
            await browser.close()
            print(f"사용자 {user_id} 시뮬레이션 완료")
import asyncio
import nest_asyncio
nest_asyncio.apply()  # 이미 실행 중인 이벤트 루프에 대한 수정

async def main():
    """40명의 사용자 시뮬레이션을 비동기 방식으로 실행"""
    tasks = [user_simulation(user_id) for user_id in range(40)]
    await asyncio.gather(*tasks)

# 기존 asyncio.run() 대신 다음 코드로 대체
loop = asyncio.get_event_loop()
loop.run_until_complete(main())
```
